```python
import datetime
import json

import pandas as pd
import pprint
```

```python
with open('data.json', 'r') as f:
    data = json.load(f)
    pprint.pprint(data[0])
```

```python
product_data = []
for product in data:
    print(product["productDetailUrl"])
    product_data.append({
        "id": product["productId"],
        "title": product["title"]["displayTitle"],
        "price": product["prices"]["salePrice"]["cent"] if "prices" in product else None,
        "url": product["productDetailUrl"],
        "tradeCount": product["trade"]["realTradeCount"],
        "starRating": product["evaluation"]["starRating"] if "evaluation" in product else None,
        "timestamp": datetime.datetime.fromisoformat(product["timestamp"])
    })
    
```

```python
df = pd.DataFrame.from_records(product_data).dropna()
df = df.sort_values('timestamp').drop_duplicates('id', keep='last')
df
```

```python
df["tradeCount"] = df["tradeCount"].astype(int)
df["metric"] = (df.tradeCount * df.starRating) / df.price
df = df.sort_values(by="metric", ascending=False)
df
```

```python
top_30 = df[df["starRating"] >= 4.8].head(30)
top_30.to_csv('products.csv', index=False) 
top_30
```
